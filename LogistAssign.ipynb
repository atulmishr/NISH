```python
#importing libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
%matplotlib inline
```

```python
#concatenate the two DSs and perform follwing basic steps:-
traind=pd.read_csv('https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/train.csv')
testd=pd.read_csv('https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/test.csv')
data=[traind,testd]
tit=pd.concat(data)
tit.head()
```

```python
tit.shape
```

```python
tit.index
```

```python
tit.columns
```

```python
#find missing values
tit.isnull().sum()
```

```python
#Removing null values
tit['Survived'] = tit['Survived'].dropna()
```

```python
tit['Survived'].value_counts()
```

```python
tit['Has_Cabin']=tit.Cabin.isnull()
```

```python
tit
```

```python
#fill the missing values via fillna
tit.isnull().sum()
```

```python
tit.Age=tit.Age.fillna(tit.Age.mean())
tit.Cabin=tit.Cabin.fillna('C85')
tit.Fare=tit.Fare.fillna(tit.Fare.mode())
tit.Embarked=tit.Embarked.fillna('C')
tit.tail()
```

```python
#Add 2new columns by Qcut.within same bin
tit['Categorical_age']=pd.qcut(tit.Age,q=5,labels=False)
tit['Categorical_fare']=pd.qcut(tit.Age,q=3,labels=False)
tit.head()

```

```python
#Dropping useless columns...name,age,tick,etc
tit=tit.drop(['Name','PassengerId','Ticket'],axis=1)
```

```python
tit=tit.drop(['Age','Fare','SibSp','Parch'],axis=1)
```

```python
tit=tit.drop(['cabin'],axis=1)
tit.head()
```

```python
tit.head()
```

```python
#transforming All varibles in numeric by using get_dummies
tit_dummies=pd.get_dummies(tit)
tit_dummies.head()
```

```python
#Specify the data columns
X = tit_dummies.drop(['Survived'], axis = 1).values
Y = tit_dummies['Survived'].values


#importing sklearn
#preparation of DSs
from sklearn.model_selection import train_test_split as tst
X_train,X_test,Y_train,Y_test=tst(X,Y,test_size=0.15)
```

```python
#Fitting the logistic regression
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,Y_train)
```

```python
#predictions
test_pred = lr.predict(X_test)
```

```python
res=pd.DataFrame({'Actual':Y_test,'Predicted':test_pred})
print(res)
```

```python
#Accurs=acy Score
from sklearn.metrics import accuracy_score
accuracy_score = lr.score(X_test,Y_test)
print("A final Accuracy Score is : {}".format(accuracy_score))
```
